### **2da Pregunta:** ¿Cuál fue la facturación por categoria em 2023?

**Observación:** Para la facturación, consideraremos aquí el producto entre el precio unitario y la cantidad, sumados al costo de envío de cada producto.


> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de barras horizontales que muestre la facturación por categoría de Zoop.

En "df", tenemos las siguientes columnas:
* categoria: categorías de los productos
* precio_unitario: precio de una unidad del producto
* cantidad: cantidad vendida del producto
* envio: costo de envío total de la compra.

Deseamos crear una visualización que muestre la facturación de cada categoría, donde la facturación se calcula multiplicando el precio unitario por la cantidad y sumando el costo de envío.

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
# Paso 1: Crear una nueva columna con la facturación de cada fila
df['facturacion'] = (df['precio_unitario'] * df['cantidad']) + df['envio']

# Paso 2: Agrupar por categoría y sumar la facturación
facturacion_por_categoria = df.groupby('categoria')['facturacion'].sum().reset_index()

# Paso 3: Ordenar de mayor a menor facturación
facturacion_por_categoria = facturacion_por_categoria.sort_values(by='facturacion', ascending=True)

# Paso 4: Visualización con barras horizontales
plt.figure(figsize=(10, 6))
sns.barplot(data=facturacion_por_categoria, x='facturacion', y='categoria', palette='crest', orient='h')

# Estética del gráfico
plt.title('Facturación por Categoría de Producto - Zoop')
plt.xlabel('Facturación Total')
plt.ylabel('Categoría de Producto')
plt.tight_layout()

# Mostrar gráfico
plt.show()
````

---

Aquí especificamos el tipo de gráfico, guiando a ChatGPT en la creación de nuestro visual. Podemos notar que en un gráfico de barras el enfoque está más centrado en las categorías, además de ser una forma más fluida de leer textos que podrían ser demasiado extensos para un gráfico de columnas.

Utiliza las bibliotecas de visualización de datos de Python para crear este gráfico.


````python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar estilo
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

# Crear gráfico de barras
ax = sns.countplot(
    x='nota',
    data=df,
    palette='viridis',  # Paleta de colores secuencial
    edgecolor='black',   # Bordes definidos
    linewidth=0.7
)

# Personalización
plt.title('Distribución de Calificaciones de Usuarios', fontsize=16, pad=20)
plt.xlabel('Calificación (0-10)', fontsize=12)
plt.ylabel('Número de Usuarios', fontsize=12)
plt.xticks(range(0, 11))  # Mostrar todas las calificaciones

# Añadir etiquetas con la frecuencia
for p in ax.patches:
    ax.annotate(
        f'{p.get_height():.0f}',
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center',
        va='center',
        xytext=(0, 5),
        textcoords='offset points',
        fontsize=9
    )

# Añadir línea de promedio
promedio = df['nota'].mean()
plt.axvline(
    x=promedio - 0.5,  # Ajuste posicional (las barras ocupan 0-10 con ancho 1)
    color='red',
    linestyle='--',
    linewidth=2,
    label=f'Promedio: {promedio:.2f}'
)

plt.legend()
plt.tight_layout()
plt.show()
````

Una vez más, no necesitamos especificar directamente el tipo de gráfico, ya que al mencionar la distribución de valores numéricos de una simple columna, ChatGPT logró interpretar y proporcionar el código para un histograma de columnas. Este gráfico es ideal para presentar la distribución de una muestra de datos o población. Cada columna aquí representa la frecuencia de repetición de un valor.

¿Te gustaría experimentar haciendo lo mismo solo con las edades?


````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configuración de estilo
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

# Crear histograma con KDE
ax = sns.histplot(
    df['edad'],
    bins=20,                  # Número de barras
    kde=True,                 # Activar curva de densidad
    color='#1f77b4',          # Color azul estándar
    edgecolor='white',        # Bordes blancos
    linewidth=0.5,            # Grosor de borde
    stat='density',           # Mostrar densidad (proporción)
    alpha=0.7                 # Transparencia
)

# Personalización de KDE
sns.kdeplot(
    df['edad'],
    color='#d62728',          # Color rojo intenso
    linewidth=2.5,            # Grosor de línea
    label='Densidad'
)

# Calcular estadísticas importantes
mean_age = df['edad'].mean()
median_age = df['edad'].median()
std_dev = df['edad'].std()

# Añadir líneas de referencia
plt.axvline(
    mean_age,
    color='#2ca02c',
    linestyle='-',
    linewidth=2,
    label=f'Media: {mean_age:.1f} años'
)

plt.axvline(
    median_age,
    color='#ff7f0e',
    linestyle='--',
    linewidth=2,
    label=f'Mediana: {median_age:.1f} años'
)

# Rango de media ± desviación estándar
plt.axvspan(
    mean_age - std_dev,
    mean_age + std_dev,
    color='gray',
    alpha=0.1,
    label=f'±1 Desv. Estándar ({std_dev:.1f} años)'
)

# Personalización del gráfico
plt.title('Distribución de Edades de Usuarios', fontsize=16, pad=20)
plt.xlabel('Edad (años)', fontsize=12)
plt.ylabel('Densidad', fontsize=12)
plt.legend(loc='upper right')

# Ajustar límites del eje X
min_age = max(0, df['edad'].min() - 5)
max_age = df['edad'].max() + 5
plt.xlim(min_age, max_age)

# Cuadrícula vertical
plt.grid(axis='y', alpha=0.3)
plt.grid(axis='x', alpha=0.1)

plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configuración de estilo
sns.set_style("whitegrid")
plt.figure(figsize=(14, 8))

# Paleta de colores
paleta = {'Masculino': '#1f77b4', 'Femenino': '#ff7f0e'}

# Crear histograma con KDE
ax = sns.histplot(
    data=df,
    x='edad',
    hue='sexo_biologico',
    bins=20,
    kde=True,
    palette=paleta,
    alpha=0.6,  # Transparencia para superposición
    edgecolor='white',
    linewidth=0.7,
    common_norm=False,  # Mostrar cada distribución por separado
    element='step'      # Bordes definidos en histograma
)

# Personalizar KDE
sns.kdeplot(
    data=df,
    x='edad',
    hue='sexo_biologico',
    palette=paleta,
    linewidth=2.5,
    common_norm=False,
    warn_singular=False
)

# Resaltar grupo dominante (30-40 años)
plt.axvspan(30, 40, color='green', alpha=0.1, label='Grupo dominante (30-40 años)')

# Calcular y marcar promedios
for sexo, color in paleta.items():
    avg = df[df['sexo_biologico'] == sexo]['edad'].mean()
    plt.axvline(
        avg,
        color=color,
        linestyle='--',
        linewidth=2,
        label=f'Promedio {sexo}: {avg:.1f} años'
    )

# Personalización
plt.title('Distribución de la Audiencia por Edad y Sexo Biológico', fontsize=18, pad=20)
plt.xlabel('Edad (años)', fontsize=14)
plt.ylabel('Densidad', fontsize=14)
plt.legend(title='Leyenda', loc='upper right', fontsize=12)

# Mejorar ejes
plt.xticks(np.arange(0, df['edad'].max()+5, 5))
plt.xlim(max(15, df['edad'].min()-5), df['edad'].max()+5)

# Añadir cuadrícula sutil
plt.grid(axis='y', alpha=0.3)
plt.grid(axis='x', alpha=0.1)

# Estadísticas en caja
props = dict(boxstyle='round', facecolor='white', alpha=0.8)
textstr = '\n'.join((
    f"Total muestras: {len(df)}",
    f"Proporción femenina: {df['sexo_biologico'].value_counts(normalize=True).get('Femenino', 0)*100:.1f}%",
    f"Edad promedio: {df['edad'].mean():.1f} años"
))
plt.text(
    0.95, 0.95, textstr,
    transform=ax.transAxes,
    verticalalignment='top',
    horizontalalignment='right',
    bbox=props,
    fontsize=12
)

plt.tight_layout()
plt.show()
````

Vamos a elegir ahora el gráfico de líneas (Ventas totales mensuales de Zoop) y aplicar una técnica más a él mediante el uso de la IA.


````python
# Paso 1: Asegurarse de que la columna 'fecha' sea datetime
df['fecha'] = pd.to_datetime(df['fecha'])

# Paso 2: Crear columna "mes_año" para agrupar por mes (con año incluido)
df['mes_año'] = df['fecha'].dt.to_period('M')

# Paso 3: Agrupar por mes_año y sumar facturación
ventas_mensuales = df.groupby('mes_año')['facturacion'].sum().reset_index()

# Paso 4: Convertir a datetime para orden cronológico y extraer nombre del mes
ventas_mensuales['mes_año'] = ventas_mensuales['mes_año'].dt.to_timestamp()
ventas_mensuales['nombre_mes'] = ventas_mensuales['mes_año'].dt.strftime('%B')

# Paso 5: Traducir los nombres de los meses
meses = {
    'January': 'Ene', 'February': 'Feb', 'March': 'Mar', 'April': 'Abr',
    'May': 'May', 'June': 'Jun', 'July': 'Jul', 'August': 'Ago',
    'September': 'Sep', 'October': 'Oct', 'November': 'Nov', 'December': 'Dic'
}
ventas_mensuales['mes_corto'] = ventas_mensuales['nombre_mes'].map(meses)

# Paso 6: Crear gráfico de líneas
plt.figure(figsize=(10, 6))
sns.lineplot(data=ventas_mensuales, x='mes_corto', y='facturacion', marker='o', sort=False)

# Estética del gráfico
plt.title('Ventas Totales Mensuales - Zoop')
plt.xlabel('Mes')
plt.ylabel('Facturación Total')
plt.grid(True)
plt.tight_layout()

# Mostrar gráfico
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# Asegurarnos que la columna 'fecha' esté en formato datetime
df['fecha'] = pd.to_datetime(df['fecha'])

# Crear columna con nombre de mes en texto
df['mes'] = df['fecha'].dt.strftime('%B')

# Diccionario con orden correcto de los meses
meses = {
    'January': 'Enero', 'February': 'Febrero', 'March': 'Marzo',
    'April': 'Abril', 'May': 'Mayo', 'June': 'Junio',
    'July': 'Julio', 'August': 'Agosto', 'September': 'Septiembre',
    'October': 'Octubre', 'November': 'Noviembre', 'December': 'Diciembre'
}

# Agrupación de ventas por mes, reindexada en orden correcto
ventas_mensuales = df.groupby('mes')['facturacion'].sum().reindex(meses.keys())

# Crear la figura
fig, ax = plt.subplots(figsize=(12, 6), facecolor=GRIS_1)
ax.set_facecolor(GRIS_1)

# Dibujar línea
ax.plot(
    ventas_mensuales.index,
    ventas_mensuales.values,
    marker='o',
    color=AZUL_1,
    linewidth=2
)

# Título
plt.title('Ventas totales mensuales en 2023', fontsize=18, color=GRIS_5)

# Etiquetas eje X
ax.set_xticks(list(meses.keys()))
ax.set_xticklabels([meses[m][:3] for m in meses.keys()], fontsize=12, color=AZUL_1)


# Etiquetas eje Y en formato "MXN$ 1.23 M"
ax.yaxis.set_major_formatter(
    mticker.FuncFormatter(lambda x, _: f"MXN$ {x/1e6:.2f} M")
)
plt.yticks(fontsize=12, color=AZUL_1)

# Quitar etiquetas de ejes
ax.set_xlabel('')
ax.set_ylabel('')


# Eliminar bordes superiores y derechos
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
````

````python
def ventas_mensuales():

  import pandas as pd
  import matplotlib.pyplot as plt
  import matplotlib.ticker as mticker
  import matplotlib.patches as patches

  # Definición de colores (si no están definidos)

  # Asegurarnos que la columna 'fecha' esté en formato datetime
  df['fecha'] = pd.to_datetime(df['fecha'])

  # Crear columna con nombre de mes en texto
  df['mes'] = df['fecha'].dt.strftime('%B')

  # Diccionario con orden correcto de los meses
  meses = {
      'January': 'Enero', 'February': 'Febrero', 'March': 'Marzo',
      'April': 'Abril', 'May': 'Mayo', 'June': 'Junio',
      'July': 'Julio', 'August': 'Agosto', 'September': 'Septiembre',
      'October': 'Octubre', 'November': 'Noviembre', 'December': 'Diciembre'
  }

  # Agrupación de ventas por mes, reindexada en orden correcto
  ventas_mensuales = df.groupby('mes')['facturacion'].sum().reindex(meses.keys())

  # Crear la figura
  fig, ax = plt.subplots(figsize=(12, 6), facecolor=GRIS_1)
  ax.set_facecolor(GRIS_1)

  # Dibujar línea
  line, = ax.plot(
      ventas_mensuales.index,
      ventas_mensuales.values,
      marker='o',
      color=AZUL_1,
      linewidth=2,
      markersize=8,
      markerfacecolor='white',
      markeredgewidth=2
  )

  # Título
  plt.title('Ventas totales mensuales en 2023', fontsize=18, color=GRIS_5, pad=20)

  # Etiquetas eje X
  ax.set_xticks(list(meses.keys()))
  ax.set_xticklabels([meses[m][:3] for m in meses.keys()], fontsize=12, color=AZUL_1)

  # Formato eje Y
  ax.yaxis.set_major_formatter(
      mticker.FuncFormatter(lambda x, _: f"MXN$ {x/1e6:.2f} M")
  )
  plt.yticks(fontsize=12, color=AZUL_1)

  # Quitar etiquetas de ejes
  ax.set_xlabel('')
  ax.set_ylabel('')

  # Eliminar bordes superiores y derechos
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)
  ax.spines['left'].set_color(GRIS_5)
  ax.spines['bottom'].set_color(GRIS_5)

  # Añadir cuadrícula sutil
  ax.grid(axis='y', color='white', linestyle='-', linewidth=1.5)

  # --- NUEVO: Añadir círculos rojos y numeración ---
  # Coordenadas de los puntos específicos (5º, 8º, 11º, 12º)
  puntos_destacados = [4, 7, 10, 11]  # Índices base 0: 5to=4, 8vo=7, 11vo=10, 12vo=11

  # Dibujar círculos rojos transparentes
  for i, punto_idx in enumerate(puntos_destacados):
      # Obtener coordenadas del punto
      x_coord = punto_idx
      y_coord = ventas_mensuales.values[punto_idx]

      # Crear círculo rojo
      circulo = patches.Circle(
          (x_coord, y_coord),
          radius=200000,  # Radio en unidades de datos (200 mil pesos)
          edgecolor=ROJO_1,
          facecolor='none',
          linewidth=2,
          linestyle='--',
          alpha=0.8
      )
      ax.add_patch(circulo)

      # Añadir numeración a la izquierda del círculo
      ax.text(
          x_coord - 0.4,  # Desplazamiento a la izquierda
          y_coord,
          str(i+1),       # Numeración (1 al 4)
          fontsize=14,
          fontweight='bold',
          color=ROJO_1,
          verticalalignment='center',
          bbox=dict(facecolor=GRIS_1, alpha=0.8, edgecolor='none', pad=2)
      )

  # Ajustar límites para espacio extra
  ax.set_xlim(-0.5, len(ventas_mensuales)-0.5)
  ax.set_ylim(bottom=0, top=ventas_mensuales.max() * 1.2)

  plt.tight_layout()
  return plt
````

````python
def ventas_por_trimestre():
  import pandas as pd
  import matplotlib.pyplot as plt
  import matplotlib.ticker as mtick

  # 1. Crear columna 'trimestre'
  df['trimestre'] = df['fecha'].dt.to_period('Q')

  # 2. Calcular facturación por trimestre y método de pago
  df['facturacion'] = (df['precio_unitario'] * df['cantidad']) + df['envio']
  ventas_trimestre = df.groupby(['trimestre', 'metodo_pago'])['facturacion'].sum().unstack()

  # 3. Crear gráfico
  fig, ax = plt.subplots(figsize=(10, 6))
  fig.patch.set_facecolor(GRIS_1)
  ax.set_facecolor(GRIS_1)

  # 4. Dibujar barras apiladas con paleta personalizada
  ventas_trimestre.plot(
      kind='bar',
      stacked=True,
      ax=ax,
      color=[BLANCO, GRIS_3, AQUA_1, AQUA_3]  # Según cantidad de métodos de pago
  )

  # 5. Personalización de título
  ax.set_title("Ventas por Trimestre y Método de Pago en 2023", fontsize=18, color=GRIS_5)

  # 6. Suprimir labels de ejes
  ax.set_xlabel("")
  ax.set_ylabel("")

  # 7. Etiquetas del eje x
  ax.set_xticklabels([str(label) for label in ventas_trimestre.index], fontsize=12, color=AZUL_1, rotation=0)

  # 8. Etiquetas eje y en formato “MXN$ X M”
  def millones(x, pos):
      return f"MXN$ {int(x/1e6)} M"
  ax.yaxis.set_major_formatter(mtick.FuncFormatter(millones))
  ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)

  # 9. Suprimir todos los bordes
  for spine in ['top', 'bottom', 'left', 'right']:
      ax.spines[spine].set_visible(False)

  # 10. Cuadrícula punteada en eje y
  ax.yaxis.grid(True, linestyle='--', alpha=0.6)
  ax.xaxis.grid(False)

  # 11. Leyenda
  ax.legend(title="Método de Pago por trimestre de 2023")

  plt.tight_layout()
  return plt
````
