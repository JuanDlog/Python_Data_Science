# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

Link archivo csv: https://github.com/alura-es-cursos/2063-visualizacion-de-datos-con-bibliotecas-de-python/blob/main/inmigrantes_canada.csv

```python
import pandas as pd

df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais', inplace=True)
df.head()
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
colombia = df.loc['Colombia', anos]
colombia
```

```python
col_dict = {'Año': colombia.index.tolist(),
             'Inmigrantes': colombia.values.tolist()}

col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2013'])
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2013'])
plt.title('Inmigración de colombianos hacia Canadá')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.show()
```

```python
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Inmigración de los 4 mayores países sudamericanos\nhacia Canada desde 1980 hasta 2013')

axs[0, 0].plot(df.loc['Colombia', anos])
axs[0, 0].set_title('Colombia')

axs[0, 1].plot(df.loc['Brasil', anos])
axs[0, 1].set_title('Brasil')

axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

axs[1, 1].plot(df.loc['Perú', anos])
axs[1, 1].set_title('Perú')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Año')
  ax.set_ylabel('Número de Inmigrantes')
  ax.grid()

plt.show()
```

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
```

```python
datos_tiendas
```

```python
fig, taxs = plt.subplots(2, 2, figsize=(14, 8))
fig.subplots_adjust(hspace=0.5, wspace=0.2)
fig.suptitle('Ventas en el período de enero a diciembre de 2022 en las tiendas A, B, C y D')

taxs[0, 0].plot(datos_tiendas['A'])
taxs[0, 0].set_title('Ventas en la tienda A')
taxs[0, 1].plot(datos_tiendas['B'])
taxs[0, 1].set_title('Ventas en la tienda B')
taxs[1, 0].plot(datos_tiendas['C'])
taxs[1, 0].set_title('Ventas en la tienda C')
taxs[1, 1].plot(datos_tiendas['D'])
taxs[1, 1].set_title('Ventas en la tienda D')

ymin = 0
ymax = 400

for ax in taxs.ravel():
  ax.set_ylim(ymin, ymax)

for ax in taxs.flat:
  ax.set_xlabel('Meses')
  ax.set_ylabel('Ventas')
  ax.grid()

plt.show()
```

```python
sudamerica = df.query('Region == "América del Sur"')
sudamerica.head()
```

```python
colores = ['royalblue', 'orange', 'forestgreen', 'orchid',
           'purple', 'brown', 'slateblue', 'gray', 'olive',
           'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12,5))

ax.bar(sudamerica.index, sudamerica['Total'], color=colores)
ax.set_title('Inmigración de sudamericanos hacia Canadá\ndurante el periodo de 1980 hasta 2013',
             loc='left', fontsize=18)
ax.set_xlabel('')
ax.set_ylabel('Numero de inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
colores = ['royalblue', 'orange', 'forestgreen', 'orchid',
           'purple', 'brown', 'slateblue', 'gray', 'olive',
           'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12,5))

ax.barh(sudamerica.index, sudamerica['Total'], color=colores)
ax.set_title('Inmigración de sudamericanos hacia Canadá\ndurante el periodo de 1980 hasta 2013',
             loc='left', fontsize=18)
ax.set_ylabel('')
ax.set_xlabel('Numero de inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
sudamerica_sorted = sudamerica.sort_values('Total', ascending=True)

colores = ['royalblue', 'orange', 'forestgreen', 'orchid',
           'purple', 'brown', 'slateblue', 'gray', 'olive',
           'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12,5))

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Inmigración de sudamericanos hacia Canadá\ndurante el periodo de 1980 hasta 2013',
             loc='left', fontsize=18)
ax.set_ylabel('')
ax.set_xlabel('Numero de inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
colores = []
for pais in sudamerica_sorted.index:
  if pais == 'Colombia':
    colores.append('forestgreen')
  else:
      colores.append('silver')

fig, ax = plt.subplots(figsize=(10,3))

ax.set_frame_on(False)
ax.xaxis.set_visible(False)
ax.tick_params(axis='y', which='both', size=0 )

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Sudamerica: Colombia es el segundo país con mas migrantes\nhacia Canadá durante el periodo de 1980 - 2013',
             loc='left', fontsize=18)
ax.set_ylabel('')
ax.yaxis.set_tick_params(labelsize=10)
# ax.set_xlabel('Numero de inmigrantes', fontsize=13)
# ax.xaxis.set_tick_params(labelsize=12)


for i, j in enumerate(sudamerica_sorted['Total']):
  ax.text(j+20,i,str(j), color='black', fontsize=8.5, ha='left', va='center')

fig.savefig('/content/inmigrantes_sudamericanos_canada.png',
            transparent=False, dpi=300, bbox_inches='tight')


plt.show()
```

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}

datos_tiendas = pd.DataFrame(ventas_2022, index=tiendas)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(13,8))
plt.subplots_adjust(hspace=0.4, wspace=0.3)

fig.suptitle('Ventas en el periodo de enero a diciembre de 2022 en las tiendas A, B, C, D', fontsize=20)

colores= ['darkviolet', 'green', 'darkblue', 'coral']

for i, ax in enumerate(ax.flat):
  ax.plot(datos_tiendas.loc[datos_tiendas.index[i]], color=colores[i], lw=3)
  ax.set_title(f'Ventas en la tienda {datos_tiendas.index[i]}', loc='left', fontsize=16)
  ax.set_xlabel('Mes', fontsize=14)
  ax.set_ylabel('Numero de ventas', fontsize=14)
  ax.tick_params(labelsize=12)
  ax.grid(color='lightgrey')




plt.show()
```
