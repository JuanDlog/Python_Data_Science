# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

Link archivo csv: https://github.com/alura-es-cursos/2063-visualizacion-de-datos-con-bibliotecas-de-python/blob/main/inmigrantes_canada.csv

```python
import pandas as pd

df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais', inplace=True)
df.head()
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
colombia = df.loc['Colombia', anos]
colombia
```

```python
col_dict = {'Año': colombia.index.tolist(),
             'Inmigrantes': colombia.values.tolist()}

col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2013'])
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2013'])
plt.title('Inmigración de colombianos hacia Canadá')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.show()
```

## Desafio Clase #1

```python
brasil = df.loc['Brasil', anos]
argentina = df.loc['Argentina', anos]
```

```python
br_dict = {'Año': brasil.index.tolist(),
           'Inmigrantes': brasil.values.tolist()}

ar_dict = {'Año': argentina.index.tolist(),
           'Inmigrantes': argentina.values.tolist()}
```

```python
datos_br = pd.DataFrame(br_dict)
datos_ar = pd.DataFrame(ar_dict)
datos_br.head()
```

```python
datos_ar.head()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(datos_br['Año'], datos_br['Inmigrantes'])
plt.plot(datos_ar['Año'], datos_ar['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2015'])
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.title('Inmigración de Brasil y Argentina hacia Canada')
plt.legend(['Brasil', 'Argentina'])
plt.show()
```
